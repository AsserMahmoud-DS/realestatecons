## NaN Replacement Strategy

```python
# Strategy based on data description and distributions:
# 
# Numeric features:
# - LotFrontage: Use median (skewed distribution)
# - MasVnrArea: Use 0 (most common value - no masonry)
# - GarageYrBlt: Use 0 or median (no garage = no year)
# 
# Categorical features (NA means 'None' or 'No such feature'):
# - PoolQC: 'None' (no pool)
# - MiscFeature: 'None' (no misc feature)
# - Alley: 'None' (no alley access)
# - Fence: 'None' (no fence)
# - MasVnrType: 'None' (no masonry veneer)
# - FireplaceQu: 'None' (no fireplace)
# - GarageType, GarageFinish, GarageQual, GarageCond: 'None' (no garage)
# - BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2: 'None' (no basement)
# - Electrical: Use mode (only 1 missing)
```

```python
def fill_missing_values(df):
    """Fill missing values based on the feature semantics."""
    df = df.copy()
    
    # Numeric: LotFrontage -> median
    df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
    
    # Numeric: MasVnrArea -> 0 (no masonry)
    df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
    
    # Numeric: GarageYrBlt -> 0 (no garage)
    df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
    
    # Categorical: 'None' for features where NA means absence
    na_means_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType', 
                  'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 
                  'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
                  'BsmtFinType1', 'BsmtFinType2']
    for col in na_means_none:
        if col in df.columns:
            df[col] = df[col].fillna('None')
    
    # Electrical: Fill with mode (most common)
    if 'Electrical' in df.columns:
        df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
    
    return df

# Apply to train and test
train_filled = fill_missing_values(train_df)
test_filled = fill_missing_values(test_df)

# Verify no missing values remain
print('Missing values after filling (train):', train_filled.isnull().sum().sum())
print('Missing values after filling (test):', test_filled.isnull().sum().sum())
```
