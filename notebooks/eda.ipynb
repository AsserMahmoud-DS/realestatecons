# Exploratory Data Analysis - House Prices Dataset

```python
from pathlib import Path
import pandas as pd
import numpy as np

# Use project root relative to this notebook
BASE_DIR = Path('/home/amraas/projects/realestatecons')
DATA_DIR = BASE_DIR / 'data' / 'raw'

train_df = pd.read_csv(DATA_DIR / 'train.csv')
test_df = pd.read_csv(DATA_DIR / 'test.csv')
```

## Dataset Size

```python
print(f'Train shape: {train_df.shape}')
print(f'Test shape:  {test_df.shape}')
```

## Column Overview

```python
numeric_cols = train_df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = train_df.select_dtypes(include=['object']).columns.tolist()

print(f'Numeric columns: {len(numeric_cols)}')
print(f'Categorical columns: {len(categorical_cols)}')
print()
print('Columns only in train (target):', set(train_df.columns) - set(test_df.columns))
```

## Training Set - Info

```python
train_df.info()
```

## Training Set - Numerical Statistics

```python
train_df.describe().T
```

## Test Set - Numerical Statistics

```python
test_df.describe().T
```

## Missing Values - Train

```python
missing_train = train_df.isnull().sum()
missing_train = missing_train[missing_train > 0].sort_values(ascending=False)
print(f'Features with missing values: {len(missing_train)}')
print(missing_train)
```

## Missing Values - Test

```python
missing_test = test_df.isnull().sum()
missing_test = missing_test[missing_test > 0].sort_values(ascending=False)
print(f'Features with missing values: {len(missing_test)}')
print(missing_test)
```

## Target Variable: SalePrice

```python
print('SalePrice Statistics:')
print(train_df['SalePrice'].describe())
print()
print(f'Skewness: {train_df["SalePrice"].skew():.4f}')
```
